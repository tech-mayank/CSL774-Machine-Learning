```python
closed = []
country_closed = []
Open = []
country_open = []
solved = []
country_solved = []
pending = []
country_pending = []
res = {}
unres = {}
total = {}
for i in df["State"]:
    res[i] = 0
    unres[i] = 0
    total[i] = 0
cnt = 0
for i, g in df.groupby(['State','Status']):
    if(i[1] == "Closed"):
        closed.append(len(g))
        res[i[0]] += len(g)
        total[i[0]] += len(g)
        country_closed.append(i[0])
    if(i[1] == "Open"):
        Open.append(len(g))
        unres[i[0]] += len(g)
        total[i[0]] += len(g)
        country_open.append(i[0])
    if(i[1] == "Solved"):
        solved.append(len(g))
        res[i[0]] += len(g)
        total[i[0]] += len(g)
        country_solved.append(i[0])
    if(i[1] == "Pending"):
        pending.append(len(g))
        unres[i[0]] += len(g)
        total[i[0]] += len(g)
        country_pending.append(i[0])
```

```python
fig = plt.figure(figsize = (80, 30)) 
  
# creating the bar plot 
plt.bar(country_closed, closed, color ='green',  
        width = 0.8) 
plt.yticks(fontsize = 60)
plt.xticks(rotation = 60,fontsize = 40) 
plt.xlabel("States",fontsize = 40) 
plt.ylabel("No. of Closed Complaints",fontsize = 40) 
plt.title("States vs No. of closed complaints",fontsize = 80) 
plt.show()
```

```python
import nltk
```

```python
data = ""
for i in df["Customer Complaint"]:
    data = data + i + " "
```

```python
data
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
tokens = word_tokenize(data)
```

```python
tokens_without_sw = [word for word in tokens if not word in stopwords.words() and word.isalnum()]
```

```python
tokens_without_sw = [word.lower() for word in tokens_without_sw]
```

```python
len(tokens_without_sw)
```

```python
freq=nltk.FreqDist(tokens_without_sw)
```

```python
freq
```

```python
freq.plot(20)
```
